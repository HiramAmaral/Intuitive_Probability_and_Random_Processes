````python
@lru_cache()
def factorial(n: int) -> int:
    '''   
    Args:
        n (int >= 0): the number to take factorial of
        
    Returns:
        (int): n!
        
    Notes: 
    
        Calculates n! recursively
        
        Uses memoization to efficiently calculate
        despite the recursive implementation.  
        
        Least Regularly Used, means that the growth 
        of the cache is limited by discarding the 
        entries that have not been read for a while.
    '''
    return 1 if n < 2 else n*factorial(n-1)

def comb(N,k):
    '''Calculates N Choose k'''
    return factorial(N) / (factorial(N-k)*factorial(k))
````

### 3.9 (t)

In this problem we see how to construct disjoint sets from ones that are not disjoint so that their unions will be the same. We consider only three sets and ask the reader to generalize the result. Calling the nondisjoint sets $A,B,C$ and the union $D = A\cup B\cup C$, we wish to find three disjoint sets $E_1,E_2,E_3$ so that $D = E_1\cup E_2 \cup E_3$. To do so let

$$
\begin{equation}
    \begin{split}
        E_1 &= A \\
        E_2 &= B - E_1 \\
        E_3 &= C - (E_1 \cup E_2)
    \end{split}
\end{equation}
$$

Using a Venn diagram explain this procedure. If we now have sets $A_1,A_2,\dots, A_N$, explain how to construct $N$ disjoint sets with the same union.

#### Answer:

- First, we select one set entirely, say $A$
- Second, we select everything in $B$ that's not in $A$ 
- Third, we select everything that's left over in $C$ that isn't in $E_1 \cup E_2$
- Lastly, we take the union of each to make $D$

To generalize this let $X = X_1\cup \dots \cup X_n $ be a union of nondisjoint sets and $X = E_1\cup \dots \cup E_n$ be a union of disjoint sets. Now we can construct $X$ through the following:

$$
\begin{equation}
    \begin{split}
        E_1 &= X_1 \\
        E_n &= X_n - \bigcup_{i=1}^{n-1}E_i
    \end{split}
\end{equation}
$$

where the first expression is the base case for the recursive formula. We can visualize this in python below. Note that the light blue color should be a part of the green set $E_2$.

### 3.15 (t)

Prove that if the sample space has size $N$, then the total number of events (including the impossible event and the certain event) is $2^N$. Hint: There are $N\choose k$ ways to choose an event with $k$ outcomes from a total of $N$ outcomes. Also, use the binomial formula:

$$
(a+b)^N = \sum_{k=0}^{N}{N\choose k}{ a^k b^{N-k}}
$$

#### Answer:

We can leverage the Binomial formula to count all the ways to choose an event $k$ from a total of $N$ outcomes:

$$
\begin{equation}
    \begin{split}
        \sum_{k=0}^{N}{N\choose k} &= \sum_{k=0}^{N}{N\choose k}{ 1^k 1^{N-k}} \\
                                                 &= {N\choose 0} + {N\choose 1} + \dots + {N\choose N} \\
                                                 &= (1+1)^N \\
                                                 &= 2^N
    \end{split}
\end{equation}
$$

A die is tossed that yields an even number with twice the probability of yielding an odd number. What is the probability of obtaining an even number, an odd number, a number that is even or odd, a number that is even and odd?

#### Answer:

The sample space remains the same, $S = \{1,2,3,4,5,6\}$. However, because of the bias in the die we need to use Axiom 2 to determine the new probabilities of each simple event:

$i:$  |  1 |  2 |  3 |  4 |  5 |  6
----|----|----|----|----|----|----
$P[i]:$| a  | 2a | a  | 2a | a  | 2a

Using Axiom 2 we see that

$$
\begin{equation}
    \begin{split}
    \sum_{i=1}^6{P[\{i\}]} &= a  + 2a + a  + 2a + a  + 2a \\
                           &= 9a \\
                           &=1
    \end{split}
\end{equation}
$$

so $a = \frac{1}{9}$.


$P[\{2,4,6\}] = P[\{2\}] + P[\{4\}] + P[\{6\}] = \frac{2}{9} + \frac{2}{9} + \frac{2}{9} = \frac{6 }{9 }$

$P[\{1,3,5\}] = P[\{1\}] + P[\{3\}] + P[\{5\}] = \frac{1}{9} + \frac{1}{9} + \frac{1}{9} = \frac{3 }{9 }$

$P[\{2,4,6\}\cup \{1,3,5\}]  = 1$

$P[\{2,4,6\}\cap \{1,3,5\}]  = 0$

Let's run a simulation to visualize this result better.

````python
#Create list of color symbols
blocks = [f"\x1b[4{i}m%s\x1b[0m" for i in range(1,7)]
nums = [f'{i}' for i in range(1,7)]
sym = [bck%n for bck,n in zip(blocks,nums)]
````

````python
p = [0.0, 1/9, 2/9, 1/9, 2/9, 1/9, 2/9]

def rand2sym(vec):
    """ Transform an array of random numbers, distributed uniformly in [0,1]
    into a sequence of symbols, chosen according to the probabilities defined by c (cumul of p)"""
    ans = []
    c = np.cumsum(p)
    counts = {i:0 for i in range(6)}
    for x in vec:
        for i in range(len(c)-1):
            if x>=c[i] and x<c[i+1]:
                ans.append(sym[i])
                counts[i] += 1
                break
    return ans, counts
````

````python
n = 1000
R = np.random.rand(n)
_syms,counts = rand2sym(R)
print(''.join(_syms),'\n')

#print counts
f = [float(y)/n for x,y in counts.items()]

for i in range(6):
    print('%s probability=%3.2f frequency= %d/%d = %3.2f '%(sym[i],p[i+1],counts[i],n,f[i]))
````

### 3.38 ðŸ˜ƒ (t)

Determine the probability that in a class of 23 students two or more students have birthdays on January 1.

#### Answer:

Let each day of the year be represented by a ball labeled as the corresponding day in the year: $\{1,2,3,\dots,365\}$. So, for instance, January 1 is ball $1$, January 2 is ball $2$, and so on. We are interested in the event 

$A = \{$ At least two students out of 23 chose ball $1\} = \{$ No students out of 23 chose ball $1\}^c$ 

We can write each outcome of this experiment as a $23-$tuple $(z_1,z_2,\dots,z_{23})$ where each $z_i$ corresponds to student $i$'s ball selection. For instance, $(z_1,z_2,\dots,z_{23}) = (1,1,\dots,1) = \mathbb{1}_{1\times23}$ is the extremely unlikely event where every student's birthday is on January 1--in fact there's only one realization where this happens!

The total number of possible birthday combinations is 

$$
N = {365 \choose 1}^{23} = 365^{23}
$$

because there are $365$ possible choices for each student's birthday (assuming equal probability and ignoring the existence of leap years).

Now the hard part: determining the size of $A$. Since we're interested in **at least two** students drawing ball $1$ that means we need to consider the cases where $k=2,3,\dots,23$. Let's start small. Say, for example, $k=2$, then two students have successfully drawn ball $1$. There are ${23 \choose 2}$ possible pairs that could have done this. Each pair had $365-364 = 1$ balls to choose from (i.e. January 1st) and the other $23-2=21$ students each had $364$ balls to choose from (i.e. every other day of the year) so, there are 

$$
{23 \choose 2}(1)^2 (364)^{23-2}
$$

possibilities. Generalizing, we see that

$$
N_A = \sum_{k=2}^{23}{ {23\choose k}\big(365-364 \big)^k 364^{23-k} }. 
$$

So the probability is 

$$
\begin{equation}
    \begin{split}
         P[k \geq 2] &= \frac{N_A}{N} \\
                     &=\frac{ \sum_{k=2}^{23}{ {23\choose k}364^{23-k} }  }{{365\choose 1}^{23}} \\
                     &\approx 0.00183.
    \end{split}
\end{equation}
$$

### 3.45 (w)

A class consists of $30$ students of which $20$ are freshman and $10$ are sophmores. If $5$ students are selected at random, what is the probability that they will all be sophmores?

#### Answer:

This is an urn problem in disguise: There are $30$ balls in an urn, $20$ red and $10$ black, and we choose $5$ at random without replacement. There are $N_S = {30 \choose 5}$ possible selections and $N_A = {10 \choose 5}$ possible ways to choose $5$ black balls:

$$
P[A] = \frac{{10 \choose 5}}{{30 \choose 5}}
$$

In a way, this is similar to the method of calculating the probability of a flush. Let's verify our analysis using python using two different methods: enumeration of all the possibilities versus combinatorics.

````python
# 1-4 are Red and 5-6 are Black
R = [*range(1,21)]
B = [*range(21,31)]
urn = R + B

# Enumerate all possibilities save for duplicates (can't choose what's already been chosen!)
possibilities = []
for v in urn:
    for w in urn:
        for x in urn:
            for y in urn:
                for z in urn:
                    if v!=w and v!=x and v!=y and v!=z and w!=x and w!=y and w!=z and x!=y and x!=z and y!=z:
                        possibilities.append((v,w,x,y,z))
                        
# Enumerate all possibilities for picking (B,B,B,B,B)
b = [pair for pair in possibilities if sum(i in B for i in pair) == 5]

N_a, N_s = len(b), len(possibilities)
# Calc the probability
prob = N_a / N_s
print(f'There are {N_a} possible ways to choose 5 black balls out of {N_s} total possible selections')
prob
````

### 3.46 (w)

An urn containing an infinite number of balls has a proportion $p$ of red balls, and the remaining $1-p$ of black balls. Two balls are chosen at random. What value of $p$ will yield the highest probability of obtaining one red ball and one black ball in any order?

#### Answer:

Here we'll need to do some simple calculus to find the argument of the maximum of a function. We are given the fact that $\frac{N_R}{N} = p$ and $\frac{N_B}{N} = \frac{N - N_R}{N} = 1-p$. Since the urn contains an extrememly large number of balls (an infinite number) the **binomial law** applies to the drawing of balls _without replacement_. This fits our problem, hence let the number of samples $m = 2$ and the number of successful draws $k=1$ (i.e. we draw one red ball). We calculate

$$
\begin{equation}
    \begin{split}
        P[k=1] &= {m\choose k}p^k(1-p)^{m-k} \\
               &= {2\choose 1}p^1(1-p)^{2-1} \\
               &= 2p - 2p^2.
    \end{split}
\end{equation}  
$$

Now we take the derivative with respect to $p$ and set the result equal to zero:

$$
\begin{equation}
    \begin{split}
        \frac{dP[k=1]}{dp} &= \frac{d}{dp}\bigg({m\choose k}p^k(1-p)^{m-k} \bigg)\bigg|_{m=2, k=1} \\
                           &= \frac{d}{dp}\big( 2p - 2p^2 \big) \\
                           &= 2 - 4p \\
                           &= 0.
    \end{split}
\end{equation}  
$$

Thus, $p=\frac{1}{2}$ will yield the highest probability of obtaining one red ball and one black ball in any order.

### 3.48 (c)

Compare the hypergeometric law to the binomial law if $N=1000, M=100, p=0.94$ by calculating the probability $P[k]$ for $k=95,96,\dots,100$. Hint: To avoid computational difficulties of calculating $N!$ for large $N$, use the following strategy to find $x=\frac{1000!}{900!}$ as an example

$$y=\ln{(x)} = \ln{(1000!)} - \ln{(900!)} = \sum_{i=1}^{1000}{\ln{(i)}} - \sum_{i=1}^{900}{\ln{(i)}}$$

and then $x=e^y$. Alternatively, for this example you can cancel out the common factors in the quotien of $x$ and write it as $x=(1000)_{100}$ which is easier to compute. But in general, this may be more difficult to set up and program.

#### Answer:

The **Hypergeometric probability law** states that 

$$
\begin{equation}
    \begin{split}
        P[k] &= \frac{ m!{N_R\choose k}{N_B\choose m-k} }{ (N)_M }\\
             &= \frac{ {N_R\choose k}{N-N_B\choose m-k} }{ {N\choose m} }.
    \end{split}
\end{equation} 
$$

In our case $N = N_R + N_B = 1000$ is the number of balls in the urn and $p = 0.94 = \frac{N_R}{N}$ is the proportion of red balls. Thus, the total number of red balls is $N_R = 940$ and total number of black balls is $N_B = N-N_R = 60$. Note that $r \leq N_R$ and $b \leq N_B$ because $r=94, b=6$ are the portion of balls observed from sampling $m=100$ balls, where $m \leq N$.

On the other hand, the **Binomial probability law** (for sampling without replacement given large $N$) states that 

$$
\begin{equation}
    \begin{split}
        P[k] = {m\choose k}p^k(1-p)^{m-k} .
    \end{split}
\end{equation} 
$$

We can actually get away with directly computing $N!$ because we used `lru_cache` memoization to effeciently compute the recursive factorial function.

````python
# Observed samples
r, b, m = 94, 6, 100

# Global numbers
N, N_R, N_B = 1000, 940, 60

# k successes (i.e. we draw k red balls)
k = [*range(95,101)]

def hypergeometric(R, B, N, m, ki):
    '''Generates a hypergeometric random sample'''
    num = comb(R, ki)*comb(B, m-ki)
    den = comb(N, m)
    return num / den

def binomial(M, p, ki):
    '''Generates a binomial random sample'''
    return comb(M, ki) * p**ki * (1-p)**(M-ki)
    
for i in range(len(k)):
    hype = hypergeometric(N_R, N_B, N, m, k[i])
    bino = binomial(m, N_R/N, k[i])
    error = abs((hype-bino)/hype)
    print(f'Hypergeometric[k={k[i]}] = {hype}\nBinomial[k={k[i]}] = {bino}\nRelative Error = {error}\n')
````

There actually exists a readily available implementation in `numpy`. For example, suppose we have an urn with 15 red and 15 black marbles. If we pull 15 marbles at random, how likely is it that 12 or more of them are one color? The answer is given by the following code snippet:
